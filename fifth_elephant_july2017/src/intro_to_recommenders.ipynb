# Standard Cases in Similarity

```python
from scipy import stats
```

```python
part_1_1 = stats.poisson.rvs(2, size=5) # Customer 1 gives lower ratings to first 5 movies / products
part_2_1 = stats.poisson.rvs(8, size=5) # Customer 1 gives higher ratings to last 5 movies / products
cust_1 = np.hstack([part_1_1, part_2_1])
print cust_1
```

```python
part_1_2 = stats.poisson.rvs(8, size=5) # Customer 2 gives higher ratings to first 5 movies / products
part_2_2 = stats.poisson.rvs(2, size=5) # Customer 2 gives lower ratings to last 5 movies / products
cust_2 = np.hstack([part_1_2, part_2_2])
print cust_2
```

```python
part_1_3 = stats.poisson.rvs(2, size=5) # Customer 3 gives lower ratings to first 5 movies / products
part_2_3 = stats.poisson.rvs(8, size=5) # Customer 3 gives higher ratings to last 5 movies / products
cust_3 = np.hstack([part_1_3, part_2_3])
print cust_3
```

```python
print distance.cosine(cust_1, cust_2)
print distance.cosine(cust_1, cust_3)
```

```python
assert(distance.cosine(cust_1, cust_2) > distance.cosine(cust_1, cust_3) )
```
